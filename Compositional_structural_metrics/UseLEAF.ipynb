# Local environment induced atomic features (LEAFs)

### Usage: [1] Load LEAF

```python
from LEAF import LEAF
```

### [2] Provide a composition as a string:

```python
comp1 = LEAF('CaTiO3')
```

 you can get a LEAFs representation of this composition:

```python
print(comp1.representation)
```

### [3] Find probability of pairwise substitution of elements, without changing the structure of this compound:

```python
print(comp1.sub_probability('Sr2Ti2O5'))
```

### [4] Find similarity between two compositions in the chemical space of inorganic materials

```python
print(comp1.similarity('SrTiO3'), comp1.similarity('SrSiO3'))
```

### [5] Find top-N substitutions, and product probability

```python
print(comp1.best_substitutions(['Ca2+'], top=7))
```
